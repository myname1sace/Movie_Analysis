### üîó Merge all tables

````python
# Merge movies and links on 'movieId'
movies_links = pd.merge(m, l, on='movieId', how='inner')

# Merge ratings and tags on 'userId' and 'movieId'
ratings_tags = pd.merge(r, t, on=['userId', 'movieId'], how='inner')

# Merge the results of the previous merges on 'movieId'
# Use suffixes to handle duplicate column names like 'timestamp'
df = pd.merge(movies_links, ratings_tags, on='movieId', how='inner', suffixes=('_movies', '_ratings_tags'))

# Display the first few rows of the merged dataframe
display(df.head())
````

## üì∫ Exploratory Data Analysis üì∫

````python
# Import the re module at the beginning of your code
import re

# Feature 1: Extract release year from title (more accurate than year_movies)
def extract_year_from_title(title):
    year_match = re.search(r'\((\d{4})\)', str(title))
    if year_match:
        return int(year_match.group(1))
    return None

df['release_year_from_title'] = df['title'].apply(extract_year_from_title)

# Feature 2: Count number of genres
df['genre_count'] = df['genres'].apply(lambda x: len(str(x).split('|')) if pd.notna(x) else 0)

# Feature 3: Movie age at time of rating
df['movie_age_at_rating'] = df['year_ratings_tags'] - df['release_year_from_title']

# Feature 4: Decade of movie release
df['release_decade'] = (df['release_year_from_title'] // 10) * 10

# Feature 5: Rating category (Low, Medium, High)
def categorize_rating(rating):
    if rating <= 2:
        return 'Low'
    elif rating <= 3.5:
        return 'Medium'
    else:
        return 'High'

df['rating_category'] = df['rating'].apply(categorize_rating)

# Feature 6: Has tag (boolean)
df['has_tag'] = df['tag'].notna()

# Feature 7: Genre diversity (unique genres per movie)
genre_lists = df.groupby('movieId')['genres'].first().apply(lambda x: len(str(x).split('|')))
df['genre_diversity'] = df['movieId'].map(genre_lists)

# Feature 8: User activity level (number of ratings per user)
user_activity = df['userId'].value_counts()
df['user_activity_level'] = df['userId'].map(user_activity)

print("\nNew features created:")
new_features = ['release_year_from_title', 'genre_count', 'movie_age_at_rating',
                'release_decade', 'rating_category', 'has_tag', 'genre_diversity', 'user_activity_level']
print(df[new_features].head())
print(f"\nTotal features now: {len(df.columns)}")
````

````python
# Set up the plotting style
plt.style.use('seaborn-v0_8')
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
fig.suptitle('MovieLens Dataset - Exploratory Data Analysis', fontsize=16, fontweight='bold')

# Insight 1: Overall rating distribution
axes[0,0].hist(df['rating'].dropna(), bins=20, alpha=0.7, color='skyblue', edgecolor='black')
axes[0,0].set_title('1. Overall Rating Distribution')
axes[0,0].set_xlabel('Rating')
axes[0,0].set_ylabel('Frequency')
axes[0,0].axvline(df['rating'].mean(), color='red', linestyle='--', label=f'Mean: {df["rating"].mean():.2f}')
axes[0,0].legend()

# Insight 2: Average rating by genre count
genre_count_ratings = df.groupby('genre_count')['rating'].mean()
axes[0,1].bar(genre_count_ratings.index, genre_count_ratings.values, color='lightgreen', alpha=0.7)
axes[0,1].set_title('2. Average Rating by Number of Genres')
axes[0,1].set_xlabel('Number of Genres')
axes[0,1].set_ylabel('Average Rating')

# Insight 3: Rating distribution by decade
decade_ratings = df.groupby('release_decade')['rating'].mean()
axes[0,2].plot(decade_ratings.index, decade_ratings.values, marker='o', linewidth=2, markersize=8)
axes[0,2].set_title('3. Average Rating by Release Decade')
axes[0,2].set_xlabel('Release Decade')
axes[0,2].set_ylabel('Average Rating')
axes[0,2].grid(True, alpha=0.3)

# Insight 4: Movie age vs rating
age_rating = df.groupby('movie_age_at_rating')['rating'].mean().head(50)  # First 50 years
axes[1,0].scatter(age_rating.index, age_rating.values, alpha=0.6, color='orange')
axes[1,0].set_title('4. Movie Age vs Average Rating')
axes[1,0].set_xlabel('Movie Age (Years)')
axes[1,0].set_ylabel('Average Rating')

# Insight 5: Top 10 most common genres
all_genres = []
for genres in df['genres'].dropna():
    all_genres.extend(genres.split('|'))
genre_counts = pd.Series(all_genres).value_counts().head(10)
axes[1,1].barh(range(len(genre_counts)), genre_counts.values, color='lightcoral')
axes[1,1].set_yticks(range(len(genre_counts)))
axes[1,1].set_yticklabels(genre_counts.index)
axes[1,1].set_title('5. Top 10 Most Common Genres')
axes[1,1].set_xlabel('Number of Movies')

# Insight 6: User activity distribution
user_activity_top = user_activity.head(20)
axes[1,2].bar(range(len(user_activity_top)), user_activity_top.values, color='purple', alpha=0.7)
axes[1,2].set_title('6. Top 20 Most Active Users')
axes[1,2].set_xlabel('User Rank')
axes[1,2].set_ylabel('Number of Ratings')

plt.tight_layout()
plt.show()

# Additional insights
print("\n" + "="*50)
print("KEY INSIGHTS SUMMARY")
print("="*50)

# Insight 1: Rating behavior
print(f"\n1. RATING BEHAVIOR:")
print(f"   - Average rating: {df['rating'].mean():.2f}/5.0")
print(f"   - Rating distribution: {df['rating_category'].value_counts().to_dict()}")

# Insight 2: Genre analysis
print(f"\n2. GENRE ANALYSIS:")
print(f"   - Average number of genres per movie: {df['genre_count'].mean():.2f}")
print(f"   - Most common genre combination: {df['genres'].mode().iloc[0] if not df['genres'].mode().empty else 'N/A'}")

# Insight 3: Temporal patterns
print(f"\n3. TEMPORAL PATTERNS:")
print(f"   - Movies span from {df['release_year_from_title'].min()} to {df['release_year_from_title'].max()}")
print(f"   - Average movie age at rating: {df['movie_age_at_rating'].mean():.1f} years")

# Insight 4: User behavior
print(f"\n4. USER BEHAVIOR:")
print(f"   - Total unique users: {df['userId'].nunique()}")
print(f"   - Most active user rated {user_activity.max()} movies")
print(f"   - Average ratings per user: {user_activity.mean():.1f}")

# Insight 5: Tag analysis
print(f"\n5. TAG ANALYSIS:")
print(f"   - Percentage of ratings with tags: {(df['has_tag'].sum() / len(df) * 100):.1f}%")
if df['has_tag'].sum() > 0:
    top_tags = df['tag'].value_counts().head(5)
    print(f"   - Top 5 tags: {list(top_tags.index)}")

# Insight 6: Movie popularity
print(f"\n6. MOVIE POPULARITY:")
movie_popularity = df['movieId'].value_counts()
print(f"   - Total unique movies: {df['movieId'].nunique()}")
print(f"   - Most rated movie has {movie_popularity.max()} ratings")
print(f"   - Average ratings per movie: {movie_popularity.mean():.1f}")

# Correlation matrix for numerical features
numerical_features = ['rating', 'genre_count', 'movie_age_at_rating', 'release_year_from_title', 'user_activity_level']
plt.figure(figsize=(10, 8))
correlation_matrix = df[numerical_features].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, square=True)
plt.title('Correlation Matrix of Numerical Features')
plt.tight_layout()
plt.show()
````

````python
# Dive 1: Rating patterns over time
plt.figure(figsize=(12, 6))
rating_trend = df.groupby('year_ratings_tags')['rating'].mean()
plt.plot(rating_trend.index, rating_trend.values, marker='o', linewidth=2)
plt.title('Average Rating Trend Over Years')
plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.grid(True, alpha=0.3)
plt.show()

# Dive 2: Genre combination analysis
genre_combinations = df['genres'].value_counts().head(10)
print("\nTop 10 Genre Combinations:")
for i, (genres, count) in enumerate(genre_combinations.items(), 1):
    print(f"{i}. {genres}: {count} ratings")

# Dive 3: User rating behavior analysis
user_rating_stats = df.groupby('userId').agg({
    'rating': ['count', 'mean', 'std']
}).round(2)
user_rating_stats.columns = ['rating_count', 'rating_mean', 'rating_std']
print(f"\nUser Rating Statistics:")
print(user_rating_stats.describe())

# Save the enhanced dataset
df.to_csv('C:/Users/USER/Downloads/HNG/cleaned_movies_data.csv', index=False)
print(f"\nCleaned dataset saved with {len(df.columns)} columns")
````
